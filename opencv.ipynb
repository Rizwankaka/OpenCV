## Read the image

```python
import cv2
import matplotlib.pyplot as plt 

# load the image inside the library 
image=cv2.imread('./image.jpeg')

# The color of the images loaded with OpenCV is the BGR (Blue-Green-Red) mode.
# But Matplotlib displays in RGB mode. So color images will be displayed incorrectly. 
# we need to convert the image from BGR to RGB mode.

# cher khani with image 

# change the colors to black & white 
image=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# RGB mode for matplotlib
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# show the image 
plt.imshow(image)
plt.show()

# save the image 
cv2.imwrite('./image2.jpeg', image)
```

## Read the live cam

```python
import cv2

cap=cv2.VideoCapture(0) # built-in camera 

while True:
    # Capture frame-by-frame
    ret, frame=cap.read()

    # Display the resulting frame
    cv2.imshow('Video', frame)
    
    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break

cap.release()
# destroy all windows 
cv2.destroyAllWindows()
```
